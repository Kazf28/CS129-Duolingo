```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import pickle
```

```python
# Set random seed
np.random.seed(42)
```

```python
# Load the data
df = pd.read_csv('Duolingo_data_03_07_final_standardized.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Basic statistics
print("\nBasic statistics:")
df.describe()
```

```python
# Select features for PCA
# Exclude 'user_id', 'churn_time', and 'churn_time_category'
feature_columns = [col for col in df.columns if col not in ['user_id', 'churn_time', 'churn_time_category']]
features = df[feature_columns]

# Display the selected features
print(f"\nSelected features for PCA: {features.columns.tolist()}")
print(f"Number of features: {features.shape[1]}")
```

```python
# Scale the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
features_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)
```

```python
# Split the data into training and testing sets (80-20 split)
X_train_df, X_test_df, y_train, y_test = train_test_split(
    features_scaled_df, 
    df['churn_time_category'], 
    test_size=0.2, 
    random_state=42
)

print(f"\nTraining set shape: {X_train_df.shape}")
print(f"Testing set shape: {X_test_df.shape}")
```

# Initial PCA to Determine Optimal Number of Components

```python
# Apply PCA initially keeping all components
pca = PCA()
X_train_pca = pca.fit_transform(X_train_df)
```

```python
# Extract eigenvalues and explained variance
eigenvalues = pca.explained_variance_
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)
```

```python
# Plot 1: Scree Plot - Eigenvalues
plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 1)
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, 'bo-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', alpha=0.5, label='Eigenvalue = 1 (Kaiser criterion)')
plt.title('Scree Plot of Eigenvalues')
plt.xlabel('Principal Components')
plt.ylabel('Eigenvalues')
plt.xticks(range(1, len(eigenvalues) + 1))
plt.grid(True, alpha=0.3)
plt.legend()
```

```python
# Plot Cumulative Variance Explained
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.6, color='skyblue', label='Individual explained variance')
plt.step(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, where='mid', label='Cumulative explained variance')
plt.axhline(y=0.8, color='r', linestyle='--', alpha=0.5, label='80% explained variance threshold')
plt.title('Explained Variance by Components')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance) + 1))
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Kaiser Criterion: Select components with eigenvalues > 1
kaiser_components = sum(eigenvalues > 1.5)
print(f"\nNumber of components with eigenvalues > 1 (Kaiser criterion): {kaiser_components}")

# Find number of components that explain at least 80% of variance
variance_components = np.argmax(cumulative_explained_variance >= 0.8) + 1
print(f"Number of components needed to explain at least 80% of variance: {variance_components}")
```

```python
# Choose the final number of components based on the Kaiser criterion
n_components = kaiser_components
print(f"\nFinal number of components selected: {n_components} (based on Kaiser criterion)")
```

# Final PCA with Selected Number of Components

```python
# Apply PCA with the optimal number of components
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_df)
X_test_pca = pca.transform(X_test_df)
```

```python
# Calculate and display the component loadings (contribution of each feature to each principal component)
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(n_components)],
    index=features.columns
)

# Display the loadings
print("\nComponent loadings:")
loadings
```

```python
# Visualize loadings with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(loadings, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('PCA Component Loadings')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(28, 7))
colors = ['#3498db', '#e74c3c']  # Blue for positive, red for negative

for i in range(n_components):
    plt.subplot(1, 4, i+1)
    
    # Sort loadings by absolute magnitude, keep top 10, and reverse order
    component_loadings = loadings[f'PC{i+1}'].sort_values(key=abs, ascending=False).head(10).iloc[::-1]
    
    bars = plt.barh(range(len(component_loadings)), component_loadings.values, 
                   color=[colors[0] if x >= 0 else colors[1] for x in component_loadings.values])
    
    plt.yticks(range(len(component_loadings)), component_loadings.index)
    plt.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
    plt.grid(axis='x', alpha=0.3)
    
    for bar in bars:
        width = bar.get_width()
        label_x_pos = width + 0.01 if width > 0 else width - 0.06
        plt.text(label_x_pos, bar.get_y() + bar.get_height()/2, f'{width:.2f}', 
                va='center', fontsize=9, color='black')
    
    plt.title(f'Principal Component {i+1} Loadings', fontsize=14)
    plt.xlabel('Loading Value', fontsize=12)
    
plt.tight_layout()
plt.subplots_adjust(wspace=0.8)
plt.show()
```

# Interpreting the Four Principal Components

## Principal Component 1: Response Time and Learning Efficiency

This component captures the speed-ability tradeoff in language learning:

Strong positive loadings for all response time variables (block, reverse translate, listen, reverse tap)
Strong negative loading for user ability score
Negative loadings for session structure variables (variance and averages)

Interpretation: PC1 represents a spectrum from efficient, capable learners (negative values) to users who take longer to respond but may be more deliberate. Users with high PC1 scores spend more time on each exercise but may have more structured, consistent learning sessions, while users with low PC1 scores have higher ability scores and complete exercises more quickly and efficiently.

## Principal Component 2: Engagement Volume and Consistency

This component shows intensity and regularity of platform usage:

All loadings are positive, suggesting a general measure of activity level
Strong positive loadings for activity metrics (blocks per day, variance measures, and session counts)
Moderate positive loadings for correctness variables

Interpretation: PC2 represents overall engagement intensity. Users with high PC2 scores engage frequently with the platform, complete more exercises per day, and have longer, more variable sessions - indicating committed, active learners who invest significant time in the app, regardless of their performance quality.

## Principal Component 3: Learning Performance and Accuracy

This component focuses on correctness and learning ability:

Strong positive loadings for correctness metrics across all exercise types
Positive loading for user ability score
Negative loadings for session structure variables

Interpretation: PC3 represents learning quality and performance. Users with high PC3 scores demonstrate better accuracy across different exercise types but may have less structured learning patterns, suggesting that high performers may engage differently than those still mastering basics.

## Principal Component 4: Learning Schedule Variability

This component reveals patterns in daily usage habits:

Very strong positive loading for sessions_per_day_variance
Strong positive loadings for daily metrics (average sessions per day, blocks per day variance)
Negative loadings for response time and correctness measures

Interpretation: PC4 represents learning schedule consistency. Users with high PC4 scores have highly variable daily usage patterns - they might study intensively on some days but skip others completely. Users with low scores follow more consistent daily routines but might spend more time per exercise with better performance.

## Overall Dimensions of User Behavior

1. How quickly they respond (PC1)
2. How much they engage (PC2)
3. How accurately they perform (PC3)
4. How consistently they schedule their learning (PC4)
